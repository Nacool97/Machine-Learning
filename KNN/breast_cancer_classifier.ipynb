```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':0.0,'B':1.0})
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.loc[:,df.columns!='diagnosis'],df['diagnosis'],random_state=0)
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1,p=1, metric="minkowski")
knn.fit(x_train,y_train)
```

```python
predict = knn.predict(x_test)
```

```python
accuracy = knn.score(x_test,y_test)
accuracy
```
